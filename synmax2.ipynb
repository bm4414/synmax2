```python
import pandas as pd
import pickle
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
voyages = pd.read_csv("voyages.csv")
voyages.head()
```

```python
predict = pd.DataFrame(columns=['vessel', 'begin_port_id', 'end_port_id','voyage'])
```

```python

input = voyages[['vessel','begin_port_id']]
output = voyages['end_port_id'].values
model = RandomForestClassifier()
model.fit(input,output)
pickle.dump(model, open('model/RF_vessel_predict.pickle', 'wb'))
for i in list(voyages['vessel'].unique()):
    vessel = voyages[voyages.vessel == i]
    output = vessel['end_port_id'].values
    last_port = [[i,output[-1]]]
    for j in [1,2,3]:
        predict_port = model.predict(last_port)
        begin_port_id = last_port[0][1]
        end_port_id = predict_port[0]
        predict = predict.append({'vessel': i, 'begin_port_id': begin_port_id, 'end_port_id': end_port_id, 'voyage': j}, ignore_index=True)
        last_port = [[i,predict_port[-1]]]
```

```python
predict
```

```python
predict.to_csv('predict.csv', index=False)
```
