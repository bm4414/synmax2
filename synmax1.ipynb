Sort tracking dataframe according to vessel ID and datetime

```python
df_tracking.sort_values(by=['vessel','datetime'], ignore_index=True, inplace = True)
df_tracking
```

Remove duplicate data and speed > 0

```python
# remove duplicate data
df_tracking = df_tracking.drop_duplicates(ignore_index=True)
df_tracking
```

```python
# remove row with the speed is not 0
df_tracking = df_tracking[df_tracking['speed'] == 0.0]
df_tracking = df_tracking.reset_index()
df_tracking
```

```python
# remove row if the lat and long is constant (but different datetime)
prev_vessel = 1
prev_lat = 0
prev_long = 0
filtered_df_tracking = pd.DataFrame(columns=['vessel', 'datetime', 'lat', 'long'])
for i, row in df_tracking.iterrows():
    if(prev_lat != row['lat'] and prev_lat != row['long']):
        filtered_df_tracking=filtered_df_tracking.append({'vessel': row['vessel'], 'datetime': row['datetime'], 'lat': row['lat'], 'long': row['long']}, ignore_index=True)
        prev_lat = row['lat']
        prev_long = row['long']
filtered_df_tracking        
```

```python
# get list of vessel
list_vessel=list(pd.unique(filtered_df_tracking['vessel']))
print("List vessel ID:",list_vessel)
```

```python
# a function to calculate squared distance
def distance(p1, p2):
    return (p2[0]-p1[0])**2+(p2[1]-p1[1])**2
```

```python
epsilon = 0.01 # squared distance tolerance
# define voyages dataframe
voyages = pd.DataFrame(columns=['vessel', 'begin_date', 'end_date', 'begin_port_id', 'end_port_id'])
for vessel_id in list_vessel:
    print("Vessel ID:", vessel_id)
    state_begin_or_end = 'begin' # reset the state
    for i, row_tracking in filtered_df_tracking.iterrows():
        # check vessel id
        if(row_tracking['vessel'] == vessel_id):
            # loop over ports
            for j, row_ports in df_ports.iterrows():
                # check distance
                p1 = (row_tracking['lat'],row_tracking['long'])
                p2 = (row_ports['lat'],row_ports['long'])
                if(distance(p1,p2) < epsilon):
                    if(state_begin_or_end == 'begin'):
                        begin_date = row_tracking['datetime']
                        begin_port_id = row_ports['port']
                        state_begin_or_end = 'end'
                    else:
                        if(row_ports['port'] != begin_port_id):
                            end_date = row_tracking['datetime']
                            end_port_id = row_ports['port']
                            voyages=voyages.append({'vessel': vessel_id, 'begin_date': begin_date, 'end_date': end_date, 'begin_port_id': begin_port_id, 'end_port_id': end_port_id}, ignore_index=True)
                            begin_date = row_tracking['datetime']
                            begin_port_id = row_ports['port']
                    
            
```
